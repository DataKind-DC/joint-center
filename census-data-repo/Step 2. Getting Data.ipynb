### Loop through all variables
Created a batch to be able to handle 50 variables at a time

```python
for i in range( 0, 44):
    lower_range = i * 50
    upper_range = min( i * 50 + 49, 2178)
    print(lower_range, upper_range)
    try:
        data_cd = censusdata.download('acs5', 2015, GEO, ID_LIST[lower_range :  upper_range], key= KEY).reset_index()
        data_cd['GEOID'] = data_cd['index'].apply(lambda x: str(x.geo[0][1])) + data_cd['index'].apply( lambda x: str(x.geo[1][1]))
        
        data_cd.to_csv('acs_2015_cd_' + str(lower_range) + '_' + str(upper_range) + '.csv')
    except:
        pass
```

### Loop through all files and marge in one single file

```python
data_cd = pd.read_csv('acs_2015_cd_0_49.csv', dtype = {'GEOID': str})
for i in range( 1, 44):
    lower_range = i * 50
    upper_range = min( i * 50 + 49, 2178)
    
    try:
        data_cd_batch = pd.read_csv('acs_2015_cd_' + str(lower_range) + '_' + str(upper_range) + '.csv', 
                                   dtype = {'GEOID': str})
        data_cd = data_cd.merge(data_cd_batch, how = 'left', on = 'GEOID')
        print(lower_range, upper_range)
    except:
        pass
```

```python
data_cd.to_csv('all_columns_black_blended.csv')
```
